```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
pp = PdfPages('kde_plots.pdf')
```

```python
generic = np.genfromtxt('events.generic.dat', dtype=float, delimiter='|', names=True)
generic['event_weight'] = generic['event_weight']*100
```

```python
from plotutils import *
```

```python
mc_evttypes = {0: 'Signal',
               1: 'Normalization',
               2: 'SLSL',
               3: 'SLHad',
               4: 'HadHad',
               5: 'Continuum'}
```

```python
fig, ax = plt.subplots(2, 2, figsize=(16,10),dpi=100)
StackByCategory(generic, 'eextra50', 'mc_evttype', 'event_weight',
                cat_dict=mc_evttypes,
                bins=100, ax=ax[0,0]);
StackByCategory(generic, 'mmiss_prime2', 'mc_evttype', 'event_weight',
                cat_dict=mc_evttypes,
                bins=100, ax=ax[0,1]);
StackByCategory(generic, 'tag_lp3', 'mc_evttype', 'event_weight',
                cat_dict=mc_evttypes,
                bins=100, ax=ax[1,0]);
StackByCategory(generic, 'R2', 'mc_evttype', 'event_weight',
                cat_dict=mc_evttypes,
                bins=100, ax=ax[1,1]);
pp.savefig(fig)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(16,10),dpi=100)
KdeOverHist(generic, 'mmiss_prime2', 'mc_evttype', 0, cat_dict=mc_evttypes,
            bandwidths=[1, 2], cv=True, ax=ax[0,0])
KdeOverHist(generic, 'eextra50', 'mc_evttype', 0, cat_dict=mc_evttypes,
            bandwidths=[0.1, 0.2], cv=True, ax=ax[0,1])
KdeOverHist(generic, 'tag_lp3', 'mc_evttype', 0, cat_dict=mc_evttypes,
            bandwidths=[0.1, 0.2], cv=True, ax=ax[1,0])
KdeOverHist(generic, 'tag_cosBY', 'mc_evttype', 0, cat_dict=mc_evttypes,
            bandwidths=[0.25, 1], cv=True, ax=ax[1,1], legend_size=10, legend_loc=2)
pp.savefig(fig)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(16,10),dpi=100)
KdeByCategory(generic, 'eextra50', 'mc_evttype', cat_dict=mc_evttypes, 
              ax=ax[0,0], exclude=[2])
KdeByCategory(generic, 'mmiss_prime2', 'mc_evttype', cat_dict=mc_evttypes, 
              ax=ax[0,1], exclude=[2])
KdeByCategory(generic, 'tag_cosBY', 'mc_evttype', cat_dict=mc_evttypes, 
              ax=ax[1,0], exclude=[2], legend_loc=2)
KdeByCategory(generic, 'sig_cosBY', 'mc_evttype', cat_dict=mc_evttypes, 
              ax=ax[1,1], exclude=[2], legend_loc=2, xlim=(-20, 5))
pp.savefig(fig)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(16,10),dpi=100)
KdeByCategory(generic, 'tag_lp3', 'mc_evttype', cat_dict=mc_evttypes, 
              ax=ax[0,0], exclude=[2])
KdeByCategory(generic, 'sig_hp3', 'mc_evttype', cat_dict=mc_evttypes, 
              ax=ax[0,1], exclude=[2])
KdeByCategory(generic, 'sig_vtxB', 'mc_evttype', cat_dict=mc_evttypes, 
              ax=ax[1,0], exclude=[2])
KdeByCategory(generic, 'sig_cosThetaDtau', 'mc_evttype', cat_dict=mc_evttypes, 
              ax=ax[1,1], exclude=[2])
pp.savefig(fig)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16,5),dpi=100)
KdeByCategory(generic, 'R2', 'mc_evttype', cat_dict=mc_evttypes, 
              ax=ax[0], exclude=[2])
KdeByCategory(generic, 'cosThetaT', 'mc_evttype', cat_dict=mc_evttypes, 
              ax=ax[1], exclude=[2], legend_loc=2)
pp.savefig(fig)
```
