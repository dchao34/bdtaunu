```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
```

```python
generic = np.genfromtxt('events.generic.dat', dtype=float, delimiter='|', names=True)
generic['event_weight'] = generic['event_weight']*100
```

```python
from plotutils.plotutils import *
```

```python
mc_evttypes = {0: r'$D\tau$',
               1: r'$D^*\tau$',
               2: r'$D\ell$',
               3: r'$D^*\ell$',
               4: r'$D^{**}$',
               5: r'$SL$',
               6: r'$Had$',
               7: r'$Cont$'}
```

```python
fig, ax = plt.subplots(2, 2, figsize=(16,10),dpi=100)
StackByCategory(generic, 'eextra50', 'mc_evttypeA', 'event_weight',
                cat_labels=mc_evttypes,
                bins=100, ax=ax[0,0]);
StackByCategory(generic, 'mmiss_prime2', 'mc_evttypeA', 'event_weight',
                cat_labels=mc_evttypes,
                bins=100, ax=ax[0,1]);
StackByCategory(generic, 'tag_lp3', 'mc_evttypeA', 'event_weight',
                cat_labels=mc_evttypes,
                bins=100, ax=ax[1,0]);
StackByCategory(generic, 'R2', 'mc_evttypeA', 'event_weight',
                cat_labels=mc_evttypes,
                bins=100, ax=ax[1,1]);
```

```python
groups = {0: [0,1], 
          1: [2,3,5], 
          2: [4], 
          3: [6], 
          4: [7]}
labels = {0: r'$D^{(*)}\tau$',
          1: r'$D^{(*)}\ell+SL$',
          2: r'$D^{**}$',
          3: r'$Had$',
          4: r'$Cont$'}
exclude = [4]
```

```python
fig, ax = plt.subplots(2, 2, figsize=(16,10),dpi=100)
StackByCategory(generic, 'eextra50', 'mc_evttypeA', 'event_weight',
                cat_groups=groups, cat_labels=labels,
                bins=100, ax=ax[0,0]);
StackByCategory(generic, 'mmiss_prime2', 'mc_evttypeA', 'event_weight',
                cat_groups=groups, cat_labels=labels,
                bins=100, ax=ax[0,1]);
StackByCategory(generic, 'tag_lp3', 'mc_evttypeA', 'event_weight',
                cat_groups=groups, cat_labels=labels,
                bins=100, ax=ax[1,0]);
StackByCategory(generic, 'R2', 'mc_evttypeA', 'event_weight',
                cat_groups=groups, cat_labels=labels,
                bins=100, ax=ax[1,1]);
```

```python
fig, ax = plt.subplots(2, 2, figsize=(16,10),dpi=100)
KdeByCategory(generic, 'eextra50', 'mc_evttypeA', 
              cat_groups=groups, cat_labels=labels, 
              ax=ax[0,0])
KdeByCategory(generic, 'mmiss_prime2', 'mc_evttypeA',  
              cat_groups=groups, cat_labels=labels, 
              ax=ax[0,1])
KdeByCategory(generic, 'tag_cosBY', 'mc_evttypeA',
              cat_groups=groups, cat_labels=labels,              
              ax=ax[1,0], legend_loc=2)
KdeByCategory(generic, 'sig_cosBY', 'mc_evttypeA',
              cat_groups=groups, cat_labels=labels,             
              ax=ax[1,1], legend_loc=2, xlim=(-20, 5))
```
