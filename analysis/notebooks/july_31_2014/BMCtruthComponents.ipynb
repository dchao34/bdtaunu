```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
pp = PdfPages('matrix.pdf')
```

```python
generic = np.genfromtxt('events.generic.dat', dtype=float, delimiter='|', names=True)
generic['event_weight'] = generic['event_weight']*1000
```

```python
from plotutils.plotutils import *
```

```python
bmctypes = {0: 'Cont',
            1: 'Dtau',
            2: 'Dstartau',
            3: 'Dl',
            4: 'Dstarl',
            5: 'Dstarstar_res',
            6: 'Dstarstar_nonres',
            7: 'D_SL',
            8: '0D_SL',
            9: '0D_Had',
            10: '1D_Had',
            11: '2D_Had',
            -1: 'undefined'}

taumctypes = {0: 'tau_e',
              1: 'tau_mu',
              2: 'tau_had', 
              -1: 'undefined'}

detail_bmctypes = {0: 'Cont',
                   1: 'Dtau_e_soft',
                   2: 'Dtau_e_hard',
                   3: 'Dtau_mu_soft',
                   4: 'Dtau_mu_hard',
                   5: 'Dtau_had_soft',
                   6: 'Dtau_had_hard',
                   7: 'Dstartau_e_soft',
                   8: 'Dstartau_e_hard',
                   9: 'Dstartau_mu_soft',
                   10: 'Dstartau_mu_hard',
                   11: 'Dstartau_had_soft',
                   12: 'Dstartau_had_hard',
                   13: 'Dl_soft',
                   14: 'Dl_hard',
                   15: 'Dstarl_soft',
                   16: 'Dstarl_hard',
                   17: 'Dstarstar_res',
                   18: 'Dstarstar_nonres',
                   19: 'D_SL',
                   20: '0D_SL',
                   21: '0D_Had',
                   22: '1D_Had',
                   23: '2D_Had',
                   -1: 'undefined'}

def DetailedBMctype(bmctype, btautype, bgamE):
    if bmctype == 0:
        return 0
    if bmctype in [5, 6, 7, 8, 9, 10, 11]:
        return bmctype + 12
    
    gamE = 0
    if bgamE > 0.1:
        gamE = 1
    
    if bmctype == 3:
        return bmctype + 10 + gamE
    if bmctype == 4:
        return bmctype + 11 + gamE
    
    tautype = 0
    if btautype < 0:
        return -1
    else:
        tautype = btautype
    
    dtautype = 0
    if bmctype == 2:
        dtautype = 1
        
    return 1 + dtautype * 6 + tautype * 2 + gamE

def CreateDetailedBMctype(arr):
    arr = arr[['b1mctype', 'b2mctype',
               'b1_taumctype', 'b2_taumctype',
               'b1_dtau_max_photon_energy', 
               'b2_dtau_max_photon_energy',
               'event_weight']]
    
    detailed_types = []
    for record in arr:
        b1mctype, b2mctype = record[0], record[1]
        b1tautype, b2tautype = record[2], record[3]
        b1gamE, b2gamE = record[4], record[5],
        event_weight = record[6]
        
        b1_detailed = DetailedBMctype(b1mctype, b1tautype, b1gamE)
        b2_detailed = DetailedBMctype(b2mctype, b2tautype, b2gamE)

        detailed_types.append((b1_detailed, b2_detailed, event_weight))
    
    detailed_arr = np.array(detailed_types, dtype=[('b1_detailed', np.int), 
                                                   ('b2_detailed', np.int),
                                                   ('event_weight', np.float)])
    return detailed_arr
```

```python
def PlotMatrix(arr, group, names):
        
    n = len(group)
    m = np.zeros((n, n))
    for record in arr[['b1_detailed', 'b2_detailed', 'event_weight']]:
        i, j = -1, -1
        b1, b2 = record[0].astype(int), record[1].astype(int)
        event_weight = record[2]
        for g in group.items():
            if b1 in g[1]:
                i = g[0]
            if b2 in g[1]:
                j = g[0]
                
        if (i == -1) or (j == -1):
            continue
        else:
            m[i,j] += event_weight
              
    fig = plt.figure(0, figsize=(15,15))
    plt.xlabel(r'truth $B_2\,MC$ type', size=20)
    plt.ylabel(r'truth $B_1\,MC$ type', size=20)
    plt.matshow(np.log10(m), 0)
    plt.xticks(np.arange(len(names)), names, size=20)
    plt.yticks(np.arange(len(names)), names, size=20)
    cbar = plt.colorbar(orientation='horizontal')
    cbar.set_label(r'$\log_{10}(Counts)$', size=20)
    cbar.ax.tick_params(labelsize=20)
    for t in cbar.ax.get_xticklabels():
        t.set_rotation(45)

    return m, fig
```

```python
d = CreateDetailedBMctype(generic)
group1 = {0: [1, 2, 3, 4, 7, 8, 9, 10],
          1: [5, 6, 11, 12],
          2: [13, 14, 15, 16],
          3: [17],
          4: [18],
          5: [21, 22, 23]}
group1_labels = [r'$D^{(*)}\tau_\ell$', 
                 r'$D^{(*)}\tau_h$', 
                 r'$D^{(*)}\ell$', 
                 r'$D^{**}_{onres}$', 
                 r'$D^{**}_{offres}$',
                 r'$Had$']


m, fig = PlotMatrix(d, group1, group1_labels) 
pp.savefig(fig)
```
